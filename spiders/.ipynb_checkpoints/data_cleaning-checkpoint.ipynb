```python
import pandas as pd
data=pd.read_excel("dataset_books_cleaned.xlsx")
data
```

```python
len(data['ISBN_13'].unique())
```

```python
data_unique=data.drop_duplicates('ISBN_13')
data_unique.loc[:,['star5','star4','star3','star2','star1']]=data_unique.loc[:,['star5','star4','star3','star2','star1']].fillna(0.00)
data_unique
```

```python

data_no_index = data_unique.reset_index(drop=True)
data_no_index['n_reviews']=data_no_index['n_reviews'].fillna(0).astype(int)
data_no_index.loc[:,['star5','star4','star3','star2','star1']]=data_no_index.loc[:,['star5','star4','star3','star2','star1']].astype(float)
data_no_index
```

```python
data_no_index.to_csv('final_book_dataset.csv')
```
